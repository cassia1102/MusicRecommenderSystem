```python
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv ('score.csv',header = None)
```

```python
df.columns = ['userID', 'itemID', 'score1','score2']
```

```python
df
```

```python
gb = df.groupby('userID')
```

```python
score_weighed = df['score1']*0.7+ df['score2']*0.3
```

```python
df['score_weighed']= score_weighed
```

```python
df_sort = df.groupby(['userID']).apply(lambda x: x.sort_values(['score_weighed'], ascending = False))
```

```python
df_sort
```

```python
a = len(df_sort)/6
```

```python
a
```

```python
df_sort.insert(5, "Predictor", [1,1,1,0,0,0]*int(a), allow_duplicates=True)
```

```python
df_sort
```

```python
df_sort['TrackID'] = df_sort['userID'].astype(str) + '_' + df_sort['itemID'].astype(str)
```

```python
df_sort
```

```python
new_df = df_sort.loc[:, ['TrackID', 'Predictor']]
```
